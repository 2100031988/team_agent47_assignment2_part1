```python
# importing necessary libraries and modules for benchmarking 

import numpy as np

from numcompute import (
    load_csv,
    StandardScaler, MinMaxScaler, OneHotEncoder, SimpleImputer,
    mean, std, quantile, histogram,   
    topk, binary_search, quickselect, 
    rank, percentile,
    accuracy, f1_score,
    grad, jacobian,                  
    Pipeline
) 
```

```python
# loading a CSV file and printing the loaded data

data = load_csv("sample.csv", delimiter=",", skip_header=1)
print("Loaded CSV:\n", data)
```

```python
# sample data with missing values to demonstrate imputation and statistics functions

X = np.array([
    [1, 2],
    [3, np.nan],
    [5, 6]
])

print("Original Data:\n", X)
```

```python
# handling missing values using SimpleImputer and then scaling the data using StandardScaler

imputer = SimpleImputer()
X_imp = imputer.fit_transform(X)


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imp)

print("Processed Data:\n", X_scaled)
```

```python
# handling missing values and then scaling the data using MinMaxScaler

mms = MinMaxScaler()
X_minmax = mms.fit_transform(X_imp)

print("MinMax Scaled:\n", X_minmax)
```

```python
# demonstrating one-hot encoding of categorical data

cat_data = np.array([
    ["A"],
    ["B"],
    ["A"]
])

encoder = OneHotEncoder()
encoded = encoder.fit_transform(cat_data)

print("One Hot Encoded:\n", encoded)
```

```python
# computing mean, median and standard deviation of the processed data using NumCompute's statistics functions

print("Mean:", mean(X_scaled))
print("Std:", std(X_scaled))
print("Median (quantile 0.5):", quantile(X_scaled, 0.5))
```

```python
# finding the top 3 values in the array using topk function and performing binary search for the value 7 in the sorted array

arr = np.array([5, 2, 9, 1, 7])

top_vals, idx = topk(arr, 3)
print("Top 3:", top_vals)

idx, found = binary_search(np.sort(arr), 7)
print("Binary search (7):", idx, found)


# quickselect (kth smallest element)

kth = quickselect(arr, 2)
print("2nd smallest (quickselect):", kth)
```

```python
# now printing the ranks and percentile for the data array

data = np.array([10, 20, 20, 30])

print("Ranks:", rank(data, method="average"))
print("Percentile 50:", percentile(data, 50))
```

```python
# here the accuracy and F1 score are computed for the true and predicted labels using metrics functions

y_true = [0, 1, 1, 0]
y_pred = [0, 1, 0, 0]

print("Accuracy:", accuracy(y_true, y_pred))
print("F1 Score:", f1_score(y_true, y_pred))
```

```python
# printing the gradient of the function f at the point (3.0, 4.0) using the grad function from NumCompute

f = lambda x: x[0]**2 + x[1]**2

g = grad(f, np.array([3.0, 4.0]))
print("Gradient:", g)

# jacobian of vector function

F = lambda x: np.array([x[0]**2, x[1]**2])
J = jacobian(F, np.array([3.0, 4.0]))

print("Jacobian:\n", J)
```

```python
import time
import numpy as np

large_arr = np.random.rand(1000000)

# Vectorized
start = time.time()
vec_mean = np.mean(large_arr)
vec_time = time.time() - start

# Loop-based
start = time.time()
s = 0
for i in large_arr:
    s += i
loop_mean = s / len(large_arr)
loop_time = time.time() - start


print("\nBenchmark Results:\n")
print(f"{'Method':<15}{'Mean':<25}{'Time (seconds)':<15}")
print("-" * 55)
print(f"{'Vectorized':<15}{vec_mean:<25}{vec_time:<15}")
print(f"{'Loop-based':<15}{loop_mean:<25}{loop_time:<15}")
```
